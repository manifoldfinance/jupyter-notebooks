````python
import requests
import json
import pandas as pd
from tqdm import tqdm
````

````python
uniswap_url = 'https://api.thegraph.com/subgraphs/name/uniswap/uniswap-v2'
sushiswap_url = 'https://api.thegraph.com/subgraphs/name/sushiswap/exchange'
````

## Top Uniswap and Sushiswap pairs by liquidity

````python
pair_query = """query{
  pairs(first: 500, orderBy: reserveUSD, orderDirection: desc) {
    id,
    token0{id,symbol,decimals},
    token1{id,symbol,decimals}
  }
}"""
````

````python
pair_req = requests.post(uniswap_url, json={'query': pair_query})
print(pair_req.status_code)
pair_data = json.loads(pair_req.text)
pair_df_data = pair_data['data']['pairs']
pair_df_uni = pd.json_normalize(pair_df_data)
pair_df_uni.head()
````

````python
pair_req = requests.post(sushiswap_url, json={'query': pair_query})
print(pair_req.status_code)
pair_data = json.loads(pair_req.text)
pair_df_data = pair_data['data']['pairs']
pair_df_sushi = pd.json_normalize(pair_df_data)
pair_df_sushi.head()
````

## Top Uniswap and Sushiswap pairs by volume

````python

pair_query = """query{
  pairs(first: 500, orderBy: volumeUSD, orderDirection: desc) {
    id,
    token0{id,symbol,decimals},
    token1{id,symbol,decimals}
  }
}"""
````

````python
pair_req = requests.post(uniswap_url, json={'query': pair_query})
print(pair_req.status_code)
pair_data = json.loads(pair_req.text)
pair_df_data = pair_data['data']['pairs']
pair_df_uni_vol = pd.json_normalize(pair_df_data)
pair_df_uni_vol.head()
````

````python
pair_req = requests.post(sushiswap_url, json={'query': pair_query})
print(pair_req.status_code)
pair_data = json.loads(pair_req.text)
pair_df_data = pair_data['data']['pairs']
pair_df_sushi_vol = pd.json_normalize(pair_df_data)
pair_df_sushi_vol.head()
````

## Merge top volume and liquidity to use only those in both

````python
uni_pairs_merged = pair_df_uni.merge(pair_df_uni_vol,on='id').drop(columns=['token0.symbol_y', 'token0.id_y', 'token0.decimals_y','token1.symbol_y', 'token1.id_y','token1.decimals_y'])
uni_pairs_merged = uni_pairs_merged.rename({'id': 'address', 'token0.id_x': 'base_address','token0.symbol_x': 'base_symbol','token0.decimals_x': 'base_decimals', 'token1.id_x': 'quote_address','token1.symbol_x': 'quote_symbol','token1.decimals_x': 'quote_decimals' }, axis=1)
# uni_pairs_merged.to_csv('uni_pairs.csv',index=False)
uni_pairs_merged.head()
````

````python

sushi_pairs_merged = pair_df_sushi.merge(pair_df_sushi_vol,on='id').drop(columns=['token0.symbol_y', 'token0.id_y', 'token0.decimals_y','token1.symbol_y', 'token1.id_y','token1.decimals_y'])
sushi_pairs_merged = sushi_pairs_merged.rename({'id': 'address', 'token0.id_x': 'base_address','token0.symbol_x': 'base_symbol','token0.decimals_x': 'base_decimals', 'token1.id_x': 'quote_address','token1.symbol_x': 'quote_symbol','token1.decimals_x': 'quote_decimals' }, axis=1)
# sushi_pairs_merged.to_csv('sushi_pairs.csv',index=False)
sushi_pairs_merged.head()
````

## Find positional ids

````python
uni_dump = pd.read_csv('../contracts/export/uniswap.v2.20211105.csv')
uni_dump['address'] = '0x'+uni_dump['address'].str.lower()
uni_dump.head()
````

````python
uni_dump_merge = uni_pairs_merged.merge(uni_dump,on='address').drop(columns=['base_symbol_y', 'base_address_y', 'base_decimals_y','quote_symbol_y', 'quote_address_y','quote_decimals_y'])
uni_dump_merge = uni_dump_merge.rename({'base_address_x': 'base_address','base_symbol_x': 'base_symbol','base_decimals_x': 'base_decimals', 'quote_address_x': 'quote_address','quote_symbol_x': 'quote_symbol','quote_decimals_x': 'quote_decimals' }, axis=1)
uni_dump_merge = uni_dump_merge[['id','address','base_address','base_symbol','base_decimals','quote_address','quote_symbol','quote_decimals']]
uni_dump_merge.to_csv('uni_pairs.csv',index=False)
uni_dump_merge.head()
````

````python
sushi_dump = pd.read_csv('../contracts/export/sushiswap.v2.20211105.csv')
sushi_dump['address'] = '0x'+sushi_dump['address'].str.lower()
sushi_dump.head()
````

````python
sushi_dump_merge = sushi_pairs_merged.merge(sushi_dump,on='address').drop(columns=['base_symbol_y', 'base_address_y', 'base_decimals_y','quote_symbol_y', 'quote_address_y','quote_decimals_y'])
sushi_dump_merge = sushi_dump_merge.rename({'base_address_x': 'base_address','base_symbol_x': 'base_symbol','base_decimals_x': 'base_decimals', 'quote_address_x': 'quote_address','quote_symbol_x': 'quote_symbol','quote_decimals_x': 'quote_decimals' }, axis=1)
sushi_dump_merge = sushi_dump_merge[['id','address','base_address','base_symbol','base_decimals','quote_address','quote_symbol','quote_decimals']]
sushi_dump_merge.to_csv('sushi_pairs.csv',index=False)
sushi_dump_merge.head()

````
